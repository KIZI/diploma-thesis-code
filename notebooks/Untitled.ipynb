```scala
import $ivy.`org.plotly-scala::plotly-render:0.8.2`
import plotly._, plotly.element._, plotly.layout._, plotly.Almond._
```

```scala
//import $ivy.`org.plotly-scala::plotly-render:0.8.1`
//import plotly._, element._, layout._, Plotly._

// Inspired from https://zwild.github.io/posts/plotly-examples-for-scala/.

val data = Seq(1, 2)

val trace1 = {
  Scatter(data, data)
    .withName("(1,1)")
}

val trace2 = {
  Scatter(data, data)
    .withName("(1,2)")
    .withXaxis(AxisReference.X2)
    .withYaxis(AxisReference.Y2)
}

val trace3 = {
  Scatter(data, data)
    .withName("(1,3)")
    .withXaxis(AxisReference.X3)
    .withYaxis(AxisReference.Y3)
}

val trace4 = {
  Scatter(data,data)
    .withName("(1,4)")
    .withXaxis(AxisReference.X4)
    .withYaxis(AxisReference.Y4)
}

val layout = 
  Layout()
    .withTitle("Multiple Custom Sized Subplots")
    .withXaxis(
      Axis()
      .withAnchor(AxisAnchor.Reference(AxisReference.Y1))
      .withDomain(Range.fromDoubleTuple(0, 0.45)))
    .withYaxis(
      Axis()
        .withAnchor(AxisAnchor.Reference(AxisReference.X1))
        .withDomain(Range.fromDoubleTuple(0.5, 1)))
    .withXaxis2(
      Axis()
        .withAnchor(AxisAnchor.Reference(AxisReference.Y2))
        .withDomain(Range.fromDoubleTuple(0.55, 1)))
    .withYaxis2(
      Axis()
        .withAnchor(AxisAnchor.Reference(AxisReference.X2))
        .withDomain(Range.fromDoubleTuple(0.8, 1)))
    .withXaxis3(
      Axis()
        .withAnchor(AxisAnchor.Reference(AxisReference.Y3))
        .withDomain(Range.fromDoubleTuple(0.55, 1)))
    .withYaxis3(
      Axis()
        .withAnchor(AxisAnchor.Reference(AxisReference.X3))
        .withDomain(Range.fromDoubleTuple(0.5, 0.75)))
    .withXaxis4(
      Axis()
        .withAnchor(AxisAnchor.Reference(AxisReference.Y4))
        .withDomain(Range.fromDoubleTuple(0, 1)))
    .withYaxis4(
      Axis()
        .withAnchor(AxisAnchor.Reference(AxisReference.X4))
        .withDomain(Range.fromDoubleTuple(0, 0.45)))


val data2 = Seq(trace1, trace2, trace3, trace4)

plot(data2,layout)
```

```scala
val data = Seq(1,2,3,2,3,2,1)
val data1 = Seq(1,2,3,3,3,2,1)
val trace1 = {Histogram(data).withName("id1")}
val trace2 = {Box(data1).withName("id2")}

val layout = Layout()
.withXaxis(Axis().withAnchor(AxisAnchor.Reference(AxisReference.Y1)).withDomain(Range.fromDoubleTuple(0, 0.45)))
.withYaxis(Axis().withAnchor(AxisAnchor.Reference(AxisReference.X1)).withDomain(Range.fromDoubleTuple(0, 1)))
.withXaxis2(Axis().withAnchor(AxisAnchor.Reference(AxisReference.Y1)).withDomain(Range.fromDoubleTuple(0.55, 1)))
.withYaxis2(Axis().withAnchor(AxisAnchor.Reference(AxisReference.X1)).withDomain(Range.fromDoubleTuple(0, 1)))

val data2 = Seq(trace1, trace2)

val subplots1 = Seq( Seq("1") , Seq("2") )
val subplots2 = Seq( Seq("id1","id2"),Seq("id1","id2") )


plot(data2,Layout().withGrid(Grid().withRows(1).withColumns(2).withSubplots(subplots1)))
```
